```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('_mpl-gallery-nogrid')


# make data
x = [0, 0]
for i in range(len(k['Survived'])):
    if k['Survived'].values[i] == 1:
        x[1] = x[1] + 1
    else:
        x[0] = x[0] + 1

colors = plt.get_cmap('Blues')(np.linspace(0.2, 0.5, 2))

# plot
fig, ax = plt.subplots(figsize=(12, 7), subplot_kw=dict(aspect="equal"), dpi= 80)
ax.pie(x, colors=colors, radius=10, center=(4, 4),
       wedgeprops={"linewidth": 1, "edgecolor": "white"}, frame=True)



plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# make data
x = [0, 0]
for i in range(len(k['Survived'])):
    if k['Survived'].values[i] == 1:
        x[1] = x[1] + 1
    else:
        x[0] = x[0] + 1

fig, ax = plt.subplots(figsize=(12, 7), subplot_kw=dict(aspect="equal"), dpi= 80)

data = x
explode = [0,0.01]

def func(pct, allvals):
    absolute = int(pct/100.*np.sum(allvals))
    return "{:.1f}% ({:d} )".format(pct, absolute)

wedges, texts, autotexts = ax.pie(data,
                                  autopct=lambda pct: func(pct, data),
                                  textprops=dict(color="w"),
                                  colors=plt.cm.Dark2.colors,
                                  startangle=140,
                                  explode=explode)

# Decoration
#ax.legend(wedges, categories, title="погибшие и выжившие", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))
plt.setp(autotexts, size=10, weight=700)
ax.set_title("Количество погибших и выживших")
plt.show()
```
