```python
import pandas as pd
df = pd.read_csv("/content/drive/My Drive/Titanic-Dataset.csv")
df

#df.head(3)
#df.tail(1)
```

```python
print(f'Количество пустых ячеек = {df.isnull().sum().sum()}')
```

```python
# Удаление всех строк содержащих хотя бы одну пустую ячейку
k = df.dropna()
k
```

```python
# Поиск уникальных значений
# df
# len(df) - len(k)

print("PassengerId =",len(k['PassengerId'].unique()))
print("Survived =",len(k['Survived'].unique()))
print("Sex =",len(k['Sex'].unique()))
print("Age =",len(k['Age'].unique()))
print("SibSp =",len(k['SibSp'].unique()))
print("Parch =",len(k['Parch'].unique()))
print("Ticket =",len(k['Ticket'].unique()))
print("Fare =",len(k['Fare'].unique()))
print("Cabin =",len(k['Cabin'].unique()))
print("Embarked =",len(k['Embarked'].unique()))
```

```python
print("PassengerId =",k['PassengerId'].value_counts())
```

```python
# Поиск макс, мин и тд значений
k.max()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.plot(k['Survived'],k['Fare'], 'o')
plt.xlabel('Выжившие')
plt.ylabel('Цена билета')
plt.title('Зависимость цены билета и выживаемости');


plt.plot(k['Survived'],k['Pclass'], 'o')
plt.xlabel('Выжившие')
plt.ylabel('Класс круиза')
plt.title('Зависимость класса круиза и выживаемости');

'''Чем дороже билет, тем больше выживаемость, хоть и не сильная корреляция,
но она есть, то же самое с классом круиза, хотя цена билета и класс скорее всего
почти одно и то же.
'''
```

```python

plt.plot(k['Survived'],k['Age'], 'o')

plt.xlabel('Выжившие')
plt.ylabel('Возраст')
plt.title('Зависимость возраста и выживаемости');

'''Выжило больше молодых людей и детей'''
```
